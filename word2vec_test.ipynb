```python
word_count = 30035
vec_len = 200
M = np.zeros((word_count, vec_len))
word_lst = []

pattern = r'^\d+\s+([a-zA-ZçğıöşüÇĞİÖŞÜ]+)\s+\[([^\]]*)]'
#pattern = r'^\d+\s+([^\d]+)\s+\[([^\]]*)]'

filename = 'word2vec.tsv'
with open(filename, 'r') as file:
    content = file.read()
    match_lst = re.finditer(pattern, content, re.MULTILINE | re.DOTALL)
    
    for counter, element in enumerate(match_lst):
        word = element.groups()[0]
        vec_str = element.groups()[1]
        vec = list(map(float, vec_str.split()))
        M[counter] = vec
        word_lst.append(word)

with open('word_lst.txt', 'w') as file:
    for word in word_lst:
        file.write('%s\n' % word)
        
np.save('word2vec', M)
```
